```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.express as px
import os

plt.style.use('seaborn-v0_8-whitegrid')
os.makedirs("../outputs/eda_figures", exist_ok=True)

# Load cleaned data
df = pd.read_csv("../outputs/amazon_cleaned.csv", low_memory=False)
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
```

```python
Q1 â€” Yearly Revenue Trend
```

```python
df['year'] = df['order_date'].dt.year
yearly = df.groupby('year', as_index=False)['final_amount_inr'].sum()
yearly.rename(columns={'final_amount_inr':'revenue'}, inplace=True)
yearly['growth_%'] = yearly['revenue'].pct_change()*100

plt.figure(figsize=(10,6))
sns.lineplot(data=yearly, x='year', y='revenue', marker='o', linewidth=2)
for i,row in yearly.iterrows():
    plt.text(row['year'], row['revenue'], f"{row['growth_%']:.1f}%", ha='center', va='bottom')
plt.title("Yearly Revenue Trend (2015â€“2025)")
plt.xlabel("Year"); plt.ylabel("Revenue (INR)")
plt.tight_layout(); plt.savefig("../outputs/eda_figures/q1_yearly_revenue.png")
plt.show()
```

```python
Q2 â€” Monthly Seasonality Heatmap
```

```python
df['month'] = df['order_date'].dt.month
pivot = df.pivot_table(values='final_amount_inr', index='month', columns='year', aggfunc='sum')

plt.figure(figsize=(12,6))
sns.heatmap(pivot, cmap='YlGnBu')
plt.title("Monthly Sales Heatmap (2015â€“2025)")
plt.xlabel("Year"); plt.ylabel("Month")
plt.tight_layout(); plt.savefig("../outputs/eda_figures/q2_monthly_heatmap.png")
plt.show()
```

```python
Q3 â€” Customer Segmentation (RFM)
```

```python
from datetime import datetime
snapshot = datetime(2026,1,1)

rfm = df.groupby('customer_id').agg({
    'order_date': lambda x: (snapshot - x.max()).days,
    'transaction_id': 'nunique',
    'final_amount_inr': 'sum'
}).rename(columns={'order_date':'Recency','transaction_id':'Frequency','final_amount_inr':'Monetary'})

rfm['R_quartile'] = pd.qcut(rfm['Recency'], 4, labels=[4,3,2,1]).astype(int)
rfm['F_quartile'] = pd.qcut(rfm['Frequency'].rank(method='first'), 4, labels=[1,2,3,4]).astype(int)
rfm['M_quartile'] = pd.qcut(rfm['Monetary'], 4, labels=[1,2,3,4]).astype(int)

plt.figure(figsize=(8,6))
sns.scatterplot(data=rfm, x='Recency', y='Monetary', size='Frequency', alpha=0.6)
plt.title("Customer Segmentation (RFM)")
plt.tight_layout(); plt.savefig("../outputs/eda_figures/q3_rfm.png")
plt.show()
```

```python
Q4 â€” Payment Method Evolution
```

```python
payment_year = df.groupby(['year','payment_method'])['final_amount_inr'].sum().reset_index()
total = payment_year.groupby('year')['final_amount_inr'].transform('sum')
payment_year['share_%'] = payment_year['final_amount_inr'] / total * 100

plt.figure(figsize=(10,6))
for method in payment_year['payment_method'].unique():
    data = payment_year[payment_year['payment_method']==method]
    plt.plot(data['year'], data['share_%'], marker='o', label=method)
plt.legend(); plt.title("Payment Method Share Evolution (2015â€“2025)")
plt.xlabel("Year"); plt.ylabel("Market Share (%)")
plt.tight_layout(); plt.savefig("../outputs/eda_figures/q4_payment_methods.png")
plt.show()
```

```python
Q8 â€” Festival Sales Impact
```

```python
import numpy as np

if 'is_festival_sale' not in df.columns:
    print("Column 'is_festival_sale' not found. Skipping analysis.")
else:
    df['is_festival_sale'] = df['is_festival_sale'].astype('bool')

    revenue_festival = df.loc[df['is_festival_sale'] == True, 'final_amount_inr'].sum()
    revenue_nonfest = df.loc[df['is_festival_sale'] == False, 'final_amount_inr'].sum()

    sizes = [revenue_festival, revenue_nonfest]
    labels = ['Festival Sales', 'Regular Sales']

    if np.isnan(sizes).any() or sum(sizes) == 0:
        print("No valid festival/non-festival data to plot.")
        print(f"Festival revenue: {revenue_festival:.2f}, Regular: {revenue_nonfest:.2f}")
    else:
        plt.figure(figsize=(6,6))
        plt.pie(
            x=sizes,
            labels=labels,
            autopct='%1.1f%%',
            startangle=90,
            colors=['gold', 'lightblue']
        )
        plt.title("Festival vs Regular Revenue Share (2015â€“2025)")
        plt.tight_layout()
        plt.savefig("../outputs/eda_figures/q8_festival_sales.png", dpi=300)
        plt.show()

```

```python
Q15 â€” Discount Effectiveness
```

```python
import os
import seaborn as sns
import matplotlib.pyplot as plt

os.makedirs("../outputs/eda_figures", exist_ok=True)

if 'discount_percent' not in df.columns:
    print("Column 'discount_percent' not found.")
else:
    # Prepare data: drop NaNs, filter unrealistic discounts
    sample_df = df[['discount_percent','final_amount_inr']].dropna()
    # ensure numeric
    sample_df['discount_percent'] = pd.to_numeric(sample_df['discount_percent'], errors='coerce')
    sample_df['final_amount_inr'] = pd.to_numeric(sample_df['final_amount_inr'], errors='coerce')
    sample_df = sample_df.dropna()
    sample_df = sample_df[(sample_df['discount_percent'] >= 0) & (sample_df['discount_percent'] <= 100)]
    
    if sample_df.empty:
        print("No valid discount data after cleaning.")
    else:
        sample_plot = sample_df.sample(n=min(5000, len(sample_df)), random_state=42)
        plt.figure(figsize=(9,6))
        sns.scatterplot(data=sample_plot, x='discount_percent', y='final_amount_inr', alpha=0.5)
        plt.title("Discount % vs Final Amount (sample)")
        plt.xlabel("Discount (%)")
        plt.ylabel("Final amount (INR)")
        out = "../outputs/eda_figures/q15_discount_scatter.png"
        plt.tight_layout()
        plt.savefig(out, bbox_inches='tight', dpi=300)
        plt.show()
        print("Saved:", out)
        sample_df['disc_bin'] = pd.cut(sample_df['discount_percent'], bins=10)
        binned = (
            sample_df.groupby('disc_bin', observed=False)['final_amount_inr']
            .agg(['mean', 'count'])
            .reset_index()
        )

        print("Binned discount summary (first 10 rows):")
        print(binned.head(10).to_string(index=False))
```

```python
Q18 â€” Product Lifecycle
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os, glob

sns.set_style("whitegrid")
os.makedirs("../outputs/eda_figures", exist_ok=True)

df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')

catalog_paths = glob.glob("../data/amazon_india_products*.csv") + glob.glob("../data/*products*.csv")

if 'launch_year' in df.columns:
    print("Using 'launch_year' present in transactions dataframe.")
    df_products = df[['product_id','launch_year']].drop_duplicates(subset=['product_id']).set_index('product_id')
    source = "transactions launch_year"
else:
    catalog_found = False
    for p in catalog_paths:
        try:
            prod = pd.read_csv(p, low_memory=False)
            if 'product_id' in prod.columns:
                catalog_found = True
                catalog_path = p
                break
        except Exception as e:
            catalog_found = False
    if catalog_found:
        print("Found product catalog:", catalog_path)
        prod_cols = prod.columns.str.lower().tolist()
        prod = prod.rename(columns={c:c.strip() for c in prod.columns})
        if 'launch_year' in prod.columns:
            df_products = prod[['product_id','launch_year']].drop_duplicates(subset=['product_id']).set_index('product_id')
            source = "catalog launch_year"
            print("Using launch_year from product catalog.")
        else:
            print("Product catalog found but no 'launch_year' column â€” will infer from sales data.")
            catalog_found = False  
    if not catalog_found:
        print("No catalog launch_year available â€” inferring launch year from first sale date.")
        first_sales = df.groupby('product_id')['order_date'].min().dropna().dt.year.rename('launch_year')
        df_products = first_sales.to_frame()
        source = "inferred from first sale"

try:
    df_products = df_products.reset_index()
except Exception as e:
    print("Unexpected error preparing product-level launch_year:", e)
    raise

df = df.merge(df_products, on='product_id', how='left', suffixes=('','_prod'))
if 'launch_year' in df.columns:
    print("Launch year merged into transactions. Sample:")
    print(df[['product_id','launch_year']].drop_duplicates().head())
else:
    print("Warning: launch_year still missing after merge. Check product_id keys.")

launch_counts = df[['product_id','launch_year']].drop_duplicates().dropna()
launch_counts['launch_year'] = launch_counts['launch_year'].astype(int)
launch_series = launch_counts.groupby('launch_year')['product_id'].nunique().sort_index()

plt.figure(figsize=(10,5))
sns.lineplot(x=launch_series.index, y=launch_series.values, marker='o')
plt.title("Product Launches per Year")
plt.xlabel("Launch Year"); plt.ylabel("Number of New Products")
out_launch = "../outputs/eda_figures/q18_product_launches.png"
plt.tight_layout()
plt.savefig(out_launch, dpi=300, bbox_inches='tight')
plt.show()
print("Saved:", out_launch)

df['year'] = df['order_date'].dt.year
active_skus = df.dropna(subset=['year']).groupby('year')['product_id'].nunique().sort_index()

plt.figure(figsize=(10,5))
sns.lineplot(x=active_skus.index, y=active_skus.values, marker='o')
plt.title("Active SKUs Sold Per Year")
plt.xlabel("Year"); plt.ylabel("Unique Products Sold")
out_active = "../outputs/eda_figures/q18_active_skus.png"
plt.tight_layout()
plt.savefig(out_active, dpi=300, bbox_inches='tight')
plt.show()
print("Saved:", out_active)

top_products = df['product_id'].value_counts().head(10).index.tolist()
sample = df[df['product_id'].isin(top_products)].copy()
sample['sale_year'] = sample['order_date'].dt.year
life = sample.groupby(['product_id','sale_year'])['transaction_id'].count().reset_index().rename(columns={'transaction_id':'sales_count'})
plt.figure(figsize=(12,6))
sns.lineplot(data=life, x='sale_year', y='sales_count', hue='product_id', marker='o')
plt.title("Sales over time for Top 10 Products (sample)")
plt.xlabel("Year"); plt.ylabel("Sales Count")
out_life = "../outputs/eda_figures/q18_sample_lifecycle.png"
plt.tight_layout()
plt.savefig(out_life, dpi=300, bbox_inches='tight')
plt.show()
print("Saved:", out_life)
```

```python
Q20 â€” Business Health Summary Dashboard
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

os.makedirs("../outputs/eda_figures", exist_ok=True)

expected_cols = ['final_amount_inr','customer_id','is_prime_member','delivery_days','product_rating']
present = {c: (c in df.columns) for c in expected_cols}
print("Column presence check:")
for c, ok in present.items():
    print(f"  {c}: {'FOUND' if ok else 'MISSING'}")

def safe_sum(col):
    return df[col].dropna().sum() if (col in df.columns) else np.nan

def safe_nunique(col):
    return int(df[col].nunique()) if (col in df.columns) else np.nan

def safe_mean(col):
    if col not in df.columns:
        return np.nan
    vals = pd.to_numeric(df[col], errors='coerce')
    return float(vals.mean()) if not vals.dropna().empty else np.nan

total_revenue = safe_sum('final_amount_inr')
active_customers = safe_nunique('customer_id')
prime_pct = safe_mean('is_prime_member') * 100 if 'is_prime_member' in df.columns else np.nan
avg_order_value = safe_mean('final_amount_inr')
on_time_pct = np.nan
if 'delivery_days' in df.columns:
    try:
        on_time_pct = (pd.to_numeric(df['delivery_days'], errors='coerce') <= 3).mean()*100
    except Exception:
        on_time_pct = np.nan
avg_rating = safe_mean('product_rating')

metrics = {
    'Total Revenue (â‚¹)': total_revenue,
    'Active Customers': active_customers,
    'Prime Members (%)': prime_pct,
    'Average Order Value (â‚¹)': avg_order_value,
    'On-Time Delivery Rate (%)': on_time_pct,
    'Average Rating': avg_rating
}

print("\n=== BUSINESS HEALTH METRICS ===")
for k, v in metrics.items():
    if pd.isna(v):
        print(f"{k}:  MISSING / N/A")
    elif isinstance(v, (int, np.integer)):
        print(f"{k}: {v:,}")
    else:
        print(f"{k}: {v:,.2f}")

numeric_for_plot = {k:v for k,v in metrics.items() if (not pd.isna(v)) and (k not in ['Average Rating'])}
if numeric_for_plot:
    plt.figure(figsize=(10,4))
    names = list(numeric_for_plot.keys())
    vals = [numeric_for_plot[n] for n in names]
    scaled_vals = vals.copy()
    bars = plt.bar(names, scaled_vals)
    for bar, val in zip(bars, vals):
        h = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, h*0.98, f"{val:,.0f}" if val>=1 else f"{val:.2f}", ha='center', va='top', color='white', fontsize=10)
    plt.title("Business Health - Key Metrics")
    plt.xticks(rotation=30, ha='right')
    plt.tight_layout()
    out_chart = "../outputs/eda_figures/q20_business_health.png"
    plt.savefig(out_chart, bbox_inches='tight', dpi=300)
    plt.show()
    print("Saved chart:", out_chart)
else:
    print("No numeric KPIs available to plot.")

```
